```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parámetros poblacionales
mu = 8  # Media poblacional
sigma = 4  # Desviación estándar poblacional
n = 30  # Tamaño de la muestra

# Error estándar de la media (SE)
SE = sigma / np.sqrt(n)

# a. Probabilidad de que la duración promedio sea menor a 7 años
x1 = 7
z1 = (x1 - mu) / SE
prob_a = norm.cdf(z1)  # P(X < 7)

# b. Probabilidad de que la duración promedio exceda 7 años
prob_b = 1 - prob_a  # P(X > 7) = 1 - P(X < 7)

# c. Probabilidad de que la duración promedio esté a no más de un año de la media poblacional (7 <= X <= 9)
x2 = 9
z2 = (x2 - mu) / SE
prob_c = norm.cdf(z2) - norm.cdf(z1)  # P(7 <= X <= 9)

# Resultados
print(f"a. P(X < 7) = {prob_a:.4f}")
print(f"b. P(X > 7) = {prob_b:.4f}")
print(f"c. P(7 <= X <= 9) = {prob_c:.4f}")
# Gráfica del error estándar (SE) vs. tamaño de muestra (n)
n_values = np.arange(1, 101)  # Tamaños de muestra de 1 a 100
SE_values = sigma / np.sqrt(n_values)  # Error estándar para cada n


```

```python

# Gráfica
plt.figure(figsize=(8, 6))
plt.plot(n_values, SE_values, marker='o', linestyle='-', color='b')
plt.title("Error estándar (SE) vs. Tamaño de muestra (n)")
plt.xlabel("Tamaño de muestra (n)")
plt.ylabel("Error estándar (SE)")
plt.grid(True)
plt.show()
```

```python


# Parámetros poblacionales
mu = 71802  # Media poblacional (salario promedio)
sigma = 4000  # Desviación estándar poblacional
n = 60  # Tamaño de la muestra

# a. Descripción de la distribución de muestreo de la media muestral X
SE = sigma / np.sqrt(n)  # Error estándar de la media
print(f"a. La distribución de muestreo de la media muestral X es aproximadamente normal con:")
print(f"   - Media (μ_X): {mu}")
print(f"   - Error estándar (SE): {SE:.2f}")

# b. Límites del promedio muestral con probabilidad 0.95
conf_level = 0.95  # Nivel de confianza
z_critical = norm.ppf((1 + conf_level) / 2)  # Valor crítico de Z para 95% de confianza
lower_limit = mu - z_critical * SE  # Límite inferior
upper_limit = mu + z_critical * SE  # Límite superior
print(f"\nb. Con probabilidad 0.95, el promedio muestral estará entre:")
print(f"   - Límite inferior: {lower_limit:.2f}")
print(f"   - Límite superior: {upper_limit:.2f}")
x = 73000
z = (x - mu) / SE
prob = 1 - norm.cdf(z)
print(f"\nc. La probabilidad de que la media muestral sea mayor a 73,000 dólares es: {prob:.4f}")
if prob < 0.05:
    print("\nd. Una media muestral de 73,000 dólares es considerada poco común (probabilidad < 0.05).")
    print("   Conclusión: Es posible que esta muestra no sea representativa de la población.")
else:
    print("\nd. Una media muestral de 73,000 dólares no es considerada poco común (probabilidad >= 0.05).")
    print("   Conclusión: No hay evidencia suficiente para considerar que la muestra no sea representativa.")
```

```python
def margen_error(n, sigma2, conf_level=0.95):
    sigma = np.sqrt(sigma2)
    z_critical = norm.ppf((1 + conf_level) / 2)
    SE = sigma / np.sqrt(n)
    return z_critical * SE

n = 30
sigma2_values = [0.2, 0.9, 1.5, 3.8]

for sigma2 in sigma2_values:
    ME = margen_error(n, sigma2)
    print(f"Margen de error para σ² = {sigma2}: {ME:.4f}")
```

```python
n_values = [50, 100, 500, 1000]
s2 = 4

for n in n_values:
    ME = margen_error(n, s2)
    print(f"Margen de error para n = {n}: {ME:.4f}")
```

```python
def margen_error_proportion(n, p=0.5, conf_level=0.95):
    SE = np.sqrt(p * (1 - p) / n)
    z_critical = norm.ppf((1 + conf_level) / 2)
    return z_critical * SE

n_values = [30, 100, 400, 1000]

for n in n_values:
    ME = margen_error_proportion(n)
    print(f"Margen de error para n = {n}: {ME:.4f}")
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Definir los subconjuntos
subsets = {
    '100': 6,   # Solo Economía
    '010': 3,   # Solo Matemáticas
    '001': 4,   # Solo Geografía
    '110': 11,  # Economía y Matemáticas
    '101': 12,  # Economía y Geografía
    '011': 10,  # Matemáticas y Geografía
    '111': 24   # Las tres materias
}

# Crear el diagrama de Venn
plt.figure(figsize=(6, 6))
venn = venn3(subsets=subsets, set_labels=('Economía', 'Matemáticas', 'Geografía'))

# Mostrar los valores dentro del diagrama
for label in venn.set_labels:
    label.set_fontsize(12)
for label in venn.subset_labels:
    if label:
        label.set_fontsize(10)

# Mostrar el gráfico
plt.title("Diagrama de Venn - Estudiantes por Materias")
plt.show()
```

```python
import numpy as np
from scipy.stats import norm

n = 985
x = 592
p_hat = x / n
se = np.sqrt(p_hat * (1 - p_hat) / n)

z = norm.ppf(0.95)

margin_error = z * se

lower = p_hat - margin_error
upper = p_hat + margin_error

print(f"Intervalo de confianza del 90%: ({lower:.4f}, {upper:.4f})")
print(f"En porcentaje: ({lower*100:.1f}%, {upper*100:.1f}%)")

if lower > 0.5:
    print("\nConclusión: La candidata republicana ganará la elección (el intervalo está por encima del 50%).")
else:
    print("\nConclusión: No se puede asegurar que la candidata ganará (el intervalo incluye o está por debajo del 50%).")
```
